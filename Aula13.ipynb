```python
#Importa a biblioteca para cálculos estatísticos do Python
import numpy as np

#Importa a biblioteca do pandas
import pandas as pd
```

```python
#Cria o array com os dados
dados = np.array([12, 15, 17, 20 ,22, 25, 28, 30, 35, 40])

#Calcula os quartis
q1 = np.percentile(dados, 25)
q2 = np.percentile(dados, 50)       #É a mediana
q3 = np.percentile(dados, 75)

#Calcula a mediana usando a própria fórmula da biblioteca
mediana = np.median(dados)          #OBS: 'mean' é a média

#Imprime os resultados
print(f'Primeiro quartil (Q1) = {q1}')
print(f'Segundo  quartil (Q2) = {q2}')
print(f'Mediana               = {mediana}')
print(f'Terceiro quartil (Q3) = {q3}')
```

```python
#Leitura de um arquivo .CSV para montar um DataFrame
pesquisa = pd.read_csv('pesquisa_satisfacao.csv', sep = ';')
pesquisa
#OBS: o tipo de dado do tempo de espera é numérico, pois ele ficou alinhado à direita
#     quando fica alinhado à esquerda, o tipo de dado é string.
```

```python
#Pega a coluna de tempo de espera e transforma em array
tempo_espera_array = np.array(pesquisa["Tempo de Espera (min)"])

#Calcula os quartis e a mediana
q1_geral = np.percentile(tempo_espera_array, 25)
q2_geral = np.percentile(tempo_espera_array, 50)
q3_geral = np.percentile(tempo_espera_array, 75)
mediana_geral = np.median(tempo_espera_array)

#Imprime os resultados
print(f'Primeiro quartil (Q1) = {q1_geral}')
print(f'Segundo  quartil (Q2) = {q2_geral}')
print(f'Mediana               = {mediana_geral}')
print(f'Terceiro quartil (Q3) = {q3_geral}')
```

```python
#Retira determinados valores da coluna "Tempo de Espera (min)" dos dados "pesquisa"
pesquisa_maiores = pesquisa.loc[pesquisa["Tempo de Espera (min)"] >= q3]
pesquisa_maiores.sort_values(by="Tempo de Espera (min)", ascending= False)
```

```python
####################################
#Atividade prática - Parte 2
####################################
pesquisa_satisfacao_alta = pesquisa.loc[pesquisa["Satisfação"] == "Alta"]
pesquisa_satisfacao_alta
```

```python
#Pega a coluna de tempo de espera e transforma em array
tempo_espera_array_satisfacao_alta = np.array(pesquisa_satisfacao_alta["Tempo de Espera (min)"])
tempo_espera_array_satisfacao_alta
```

```python
#Calcula os quartis, a mediana e a média
q1 = np.percentile(tempo_espera_array_satisfacao_alta, 25)
q2 = np.percentile(tempo_espera_array_satisfacao_alta, 50)
q3 = np.percentile(tempo_espera_array_satisfacao_alta, 75)
mediana = np.median(tempo_espera_array_satisfacao_alta)
media = np.mean(tempo_espera_array_satisfacao_alta)

#Imprime os resultados
print(f'Primeiro quartil (Q1) = {q1}')
print(f'Segundo  quartil (Q2) = {q2}')
print(f'Mediana               = {mediana}')
print(f'Terceiro quartil (Q3) = {q3}')
print(f'Média                 = {media}')

#Imprime os resultados do geral
print('**************************************')
print(f'Primeiro quartil geral (Q1) = {q1_geral}')
print(f'Segundo  quartil geral (Q2) = {q2_geral}')
print(f'Mediana geral               = {mediana_geral}')
print(f'Terceiro quartil geral (Q3) = {q3_geral}')
```
