```python
#Importa a biblioteca PANDAS
import pandas as pd

#Carrega o arquivo CSV (separador ";" e encoding "latin1")
df = pd.read_csv('Precos_01sem_comb_autom_2025.csv', sep = ';', encoding = 'latin1')
df.head(2)
```

```python
#Altera o campo "Valor de Venda" de string para float (pois estava dando erro ao calcular a média)
df['Valor de Venda'] = df['Valor de Venda'].str.replace(',','.').astype(float)
df.head(2)
```

```python
#Cria um DataFrame só com os dados do município do Rio de Janeiro
df_RJ = df.loc[df['Municipio'] == 'RIO DE JANEIRO']
df_RJ.head(2)
```

```python
#Cria um DataFrame para cada tipo de combustível
df_RJ_Etanol = df_RJ.loc[df_RJ['Produto'] == 'ETANOL']
df_RJ_Gasolina = df_RJ.loc[df_RJ['Produto'] == 'GASOLINA']
df_RJ_Diesel = df_RJ.loc[df_RJ['Produto'] == 'DIESEL']
```

```python
#Cria, para cada tipo de combustível, um DataFrame com a média do valor de combustível agrupado por bairro 
df_RJ_Etanol_Bairros = df_RJ_Etanol.groupby('Bairro')['Valor de Venda'].mean().reset_index()
df_RJ_Gasolina_Bairros = df_RJ_Gasolina.groupby('Bairro')['Valor de Venda'].mean().reset_index()
df_RJ_Diesel_Bairros = df_RJ_Diesel.groupby('Bairro')['Valor de Venda'].mean().reset_index()
```

```python
#Resultado - 1A - ETANOL - Bairros com o MENOR valor médio de venda
df_RJ_Etanol_Bairros_ASC = df_RJ_Etanol_Bairros.sort_values(by='Valor de Venda')
df_RJ_Etanol_Bairros_ASC.head()
```

```python
#Resultado - 1B - ETANOL - Bairros com o MAIOR valor médio de venda
df_RJ_Etanol_Bairros_DESC = df_RJ_Etanol_Bairros.sort_values(by='Valor de Venda', ascending= False)
df_RJ_Etanol_Bairros_DESC.head()
```

```python
#Resultado - 2A - GASOLINA - Bairros com o MENOR valor médio de venda
df_RJ_Gasolina_Bairros_ASC = df_RJ_Gasolina_Bairros.sort_values(by='Valor de Venda')
df_RJ_Gasolina_Bairros_ASC.head()
```

```python
#Resultado - 2B - GASOLINA - Bairros com o MAIOR valor médio de venda
df_RJ_Gasolina_Bairros_DESC = df_RJ_Gasolina_Bairros.sort_values(by='Valor de Venda', ascending= False)
df_RJ_Gasolina_Bairros_DESC.head()
```

```python
#Resultado - 3A - DIESEL - Bairros com o MENOR valor médio de venda
df_RJ_Diesel_Bairros_ASC = df_RJ_Diesel_Bairros.sort_values(by='Valor de Venda')
df_RJ_Diesel_Bairros_ASC.head()
```

```python
#Resultado - 3B - DIESEL - Bairros com o MAIOR valor médio de venda
df_RJ_Diesel_Bairros_DESC = df_RJ_Diesel_Bairros.sort_values(by='Valor de Venda', ascending= False)
df_RJ_Diesel_Bairros_DESC.head()
```
