```python
#Importo as bibliotecas
import pandas as pd
import numpy as np
```

```python
#Carrego a base de dados
dfGeral = pd.read_csv("https://www.ispdados.rj.gov.br/Arquivos/BaseMunicipioMensal.csv", sep=";", encoding="latin1")
dfGeral.head(2)
```

```python
#Seleciono os dados dos municípios da Região dos Lagos
#Fonte: https://sebrae.com.br/Sebrae/Portal%20Sebrae/UFs/RJ/Anexos/SebraePainel_RegiaoLagos.pdf
arrayRegiaoLagos = ["Araruama", "Armação dos Búzios", "Arraial do Cabo", "Cabo Frio",
                    "Casimiro de Abreu", "Iguaba Grande", "São Pedro da Aldeia", "Saquarema", "Silva Jardim"]
dfRegLagos = dfGeral.loc[dfGeral["fmun"].isin(arrayRegiaoLagos)]
dfRegLagos.head(2)
```

```python
#Soma os valores de crimes violentos (cvli) por município (fmun)
dfCrimesViolentosRegLagos = dfRegLagos.groupby("fmun")["cvli"].sum().reset_index()
dfCrimesViolentosRegLagos
```

```python
#Exporto os dados do Data Frame para um arquivo CSV
dfCrimesViolentosRegLagos.to_csv("dfMunicCrimesViolentosRegLagos.csv", ";", index=False)
```

```python
#Crio um array com os dados dos crimes violentos na Região dos Lagos
arrayCrimesViolentos = np.array(dfCrimesViolentosRegLagos["cvli"])
arrayCrimesViolentos
```

```python
#Cálculos estatísticos
media = np.mean(arrayCrimesViolentos)
mediana = np.median(arrayCrimesViolentos)
q1 = np.percentile(arrayCrimesViolentos, 25)
q2 = np.percentile(arrayCrimesViolentos, 50)
q3 = np.percentile(arrayCrimesViolentos, 75)
distancia = (media - mediana) / mediana
iqr = q3 - q1
limSuperior = q3 + (1.5 * iqr)
limInferior = q1 - (1.5 * iqr)

print(f'Média     = {media:.2f}')
print(f'Mediana   = {mediana:.2f}')
print(f'Q1        = {q1:.2f}')
print(f'Q2        = {q2:.2f}')
print(f'Q3        = {q3:.2f}')
print(f'Distância = {distancia * 100:.2f} %')       #Em percentual
print(f'Lim. Sup. = {limSuperior:.2f}')
print(f'Lim. Inf. = {limInferior:.2f}')
```

```python
#Verifico os Outliers
dfCrimesViolentosRegLagosOutliers = dfCrimesViolentosRegLagos.loc[dfCrimesViolentosRegLagos["cvli"] >= limSuperior]
dfCrimesViolentosRegLagosOutliers
```
