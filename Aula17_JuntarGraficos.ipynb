```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Cria o Dataframe com os dados das vendas nos dias da semana
df_dias = pd.DataFrame({
    'Dias': ['Segunda','Tera','Quarta','Quinta','Sexta'],
    'Vendas': [500, 200, 1000, 150, 1950]
})

#Cria o Dataframe com os dados das vendas dos produtos
df_produto = pd.DataFrame({
    'Produto': ['PO', 'QUEIJO', 'PRESUNTO'],
    'Vendas': [120, 90, 150]
})

#Cria uma figura com 4 reas de grficos (2 linhas x 2 colunas)
fig, axs = plt.subplots(2, 2, figsize=(12, 8))
#axs  uma matriz 2x2 de objetos Axes, onde cada posio representa um grfico
#axs[0,0] -> linha 0, coluna 0 -> canto superior esquerdo 
#axs[0,1] -> linha 0, coluna 1 -> canto superior direito
#axs[1,0] -> linha 1, coluna 0 -> canto inferior esquerdo
#axs[1,1] -> linha 1, coluna 1 -> canto inferior direito

#Grfico 1: Linha - Vendas por dia da semana
axs[0,0].plot(df_dias['Dias'], df_dias['Vendas'], marker='o', linestyle='-', color='#FF8C00')
axs[0,0].set_title('Vendas por Dia da Semana')
axs[0,0].set_xlabel('Dia da Semana')
axs[0,0].set_ylabel('Valor Vendido')

#Grfico 2: Barras Verticais - Vendas por produto
axs[0,1].bar(df_produto['Produto'], df_produto['Vendas'], color ='green')
axs[0,1].set_title('Vendas por Produto')
axs[0,1].set_xlabel('Produto')
axs[0,1].set_ylabel('Vendas')

#Grfico 3: Barras Horizontais - Vendas por produto
axs[1,0].barh(df_produto['Produto'], df_produto['Vendas'], color ='brown')
axs[1,0].set_title('Vendas por Produto (Horizontal)')
axs[1,0].set_xlabel('Produto')
axs[1,0].set_ylabel('Vendas')

#Grfico 4: Quadro de texto explicativo
axs[1,1].axis('off')    #Desativa os eixos para poder incluir texto
axs[1,1].text(0.1, 1.0, 'Resumo da Aula:', fontsize=14, color='black')                      #A posio  em percentual da tela (x, y)
axs[1,1].text(0.1, 0.8, 'Senac Rio de Janeiro', fontsize=12, color='black')                 #A posio  em percentual da tela (x, y)
axs[1,1].text(0.1, 0.6, '- subplots usa matriz de eixos', fontsize=12, color='black')       #A posio  em percentual da tela (x, y)
axs[1,1].text(0.1, 0.5, '- mais controle sobre cada grfico', fontsize=12, color='black')   #A posio  em percentual da tela (x, y)
axs[1,1].text(0.1, 0.4, '- ideal para projetos mais complexos', fontsize=12, color='black') #A posio  em percentual da tela (x, y)

#Ajusta os espaos entre os grficos para evitar sobreposio
plt.tight_layout()

#Exibe todos os grficos
plt.show()
```
