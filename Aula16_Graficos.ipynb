```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({
    'Ano': [2020, 2021, 2022, 2023],    #Dicionário com os valores 
    'Vendas': [100, 150, 130, 180]      #Dicionário com os valores
})

#OBS: Passo a coluna ANO para STRING para não aparecer no gráfco 2020.0, 2020.5, 2021.0, 2021.5...
df['Ano'] = df['Ano'].astype(str)

plt.plot(df['Ano'], df['Vendas'])       #Define os dados das colunas X e Y (respectivamente)
plt.title('Vendas por Ano')             #Define o título do gráfico
plt.xlabel('Ano')                       #Define o título do eixo X
plt.ylabel('Vendas')                    #Define o título do eixo Y
plt.grid(True)                          #Exibe ou não as linhas de grade no gráfico
plt.show()                              #Exibe o gráfico
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({
    'Produto': ['A', 'B', 'C'],         #Dicionário com os valores 
    'Vendas': [120, 90, 150]            #Dicionário com os valores
})

#Ordena o gráfico primeiro
#quando coloca o "inplace=True" não precisa colocar o "df ="
#df = df.sort_values(by= 'Vendas', ascending=False)
df.sort_values(by= 'Vendas', ascending=False, inplace=True)

plt.bar(df['Produto'], df['Vendas'])    #Define os dados das colunas X e Y (respectivamente)
plt.title('Vendas por Produto')         #Define o título do gráfico
plt.xlabel('Produto')                   #Define o título do eixo X
plt.ylabel('Vendas')                    #Define o título do eixo Y
plt.grid(False)                         #Exibe ou não as linhas de grade no gráfico
plt.show()    
```

```python
import matplotlib.pyplot as plt
import pandas as pd

#Cria os dicionários com os dados
dados = {
    'Ano': [2018, 2019, 2020, 2021, 2022],
    'Vendas': [150, 200, 250, 300, 350]
}

#Seta a variável com os dicionários
df = pd.DataFrame(dados)

plt.figure(figsize=(8,5))                           #Determina o tamanho do gráfico (X e Y, respectivamente)
plt.barh(df['Ano'], df['Vendas'], color='blue')     #Define os dados das colunas X e Y (respectivamente)
plt.title('Vendas por Ano')                         #Define o título do gráfico
plt.xlabel('Vendas')                                #Define o título do eixo X
plt.ylabel('Ano')                                   #Define o título do eixo Y
plt.grid(False)                                     #Exibe ou não as linhas de grade no gráfico
plt.savefig('grafico_barras_horizontais.png')       #Salva o gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

#Cria os dicionários com os dados
df = {
    'Dias': ['Segunda','Terça','Quarta','Quinta','Sexta'],
    'Vendas': [500, 200, 1000, 150, 1950]
}

dias = df['Dias']
valores = df['Vendas']

plt.plot(dias, valores, marker='o', linestyle='-', color='green')
plt.title('Vendas por Dias')
plt.xlabel('Dia da Semana')
plt.ylabel('Valor Vendido')
plt.grid(False)
plt.show()
```
