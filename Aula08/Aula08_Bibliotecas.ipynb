```python
#Seleciona linhas dentro do arquivo passado um parâmetro
df_rj = df.loc[df['munic'] == 'Rio de Janeiro']
df_rj
```

```python
#Ordenar por uma coluna (padrão é ordenar de forma crescente)
df_rj_ordenado = df_rj.sort_values(by='pessoas_desaparecidas')
df_rj_ordenado

df_rj_ordenado = df_rj.sort_values(by='pessoas_desaparecidas', ascending= False)
df_rj_ordenado
```

```python
#Agrupamento de dados   
df_regiao_ocorrencias = df.groupby('regiao')['registro_ocorrencias'].sum().reset_index()
df_regiao_ocorrencias
```
