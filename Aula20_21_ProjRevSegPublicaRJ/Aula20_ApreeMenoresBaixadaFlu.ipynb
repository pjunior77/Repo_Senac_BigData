```python
#Importo as bibliotecas
import pandas as pd
import numpy as np
```

```python
#Carrego a base de dados
dfGeral = pd.read_csv("https://www.ispdados.rj.gov.br/Arquivos/BaseMunicipioMensal.csv", sep=";", encoding="latin1")
dfGeral.head(2)
```

```python
#Seleciono os dados da Baixada Fluminense
dfBaixFlu = dfGeral.loc[dfGeral["regiao"] == "Baixada Fluminense"]
dfBaixFlu.head(2)
```

```python
#Soma os valores de apreensão de menores (aaapai) por município (fmun)
dfApreeMenores = dfBaixFlu.groupby("fmun")["aaapai"].sum().reset_index()
#dfApreeMenores.head(2)
dfApreeMenores
```

```python
#Cria o array com os dados da apreensão de menores
arrayApreeMenores = np.array(dfApreeMenores["aaapai"])
arrayApreeMenores
```

```python
#Cálculos estatísticos
media = np.mean(arrayApreeMenores)
mediana = np.median(arrayApreeMenores)
q1 = np.percentile(arrayApreeMenores, 25)
q2 = np.percentile(arrayApreeMenores, 50)
q3 = np.percentile(arrayApreeMenores, 75)
distancia = (media - mediana) / mediana
iqr = q3 - q1
limSuperior = q3 + (1.5 * iqr)
limInferior = q1 - (1.5 * iqr)
variancia = np.var(arrayApreeMenores)
desvio_padrao = np.std(arrayApreeMenores)
coeficiente_variacao = (desvio_padrao / media) * 100
distancia_variancia = (variancia / (media ** 2)) * 100

print(f'Média     = {media:.2f}')
print(f'Mediana   = {mediana:.2f}')
print(f'Q1        = {q1:.2f}')
print(f'Q2        = {q2:.2f}')
print(f'Q3        = {q3:.2f}')
print(f'Distância = {distancia * 100:.2f} %')       #Em percentual
print(f'Lim. Sup. = {limSuperior:.2f}')
print(f'Lim. Inf. = {limInferior:.2f}')
print("Variância:", round(variancia, 2))
print("Desvio padrão:", round(desvio_padrao, 2))
print("Coeficiente de variação:", round(coeficiente_variacao, 2))
print("Distância da variância em relação à média:", round(distancia_variancia, 2))
```

```python
#Crio um dataFrame com os valores >= q3 (25% superior)
df25Mais = dfApreeMenores.loc[dfApreeMenores["aaapai"] >= q3]
df25Mais
```

```python
#Exporto o DataFrame para um arquivo CSV
df25Mais.to_csv("dfMunic25MaisApreeMenoresBaixadaFlu.csv", ";", index=False)
```

```python
#Crio um dataFrame com os valores >= lim. Superior (Outliers)
dfOutliers = dfApreeMenores.loc[dfApreeMenores["aaapai"] >= limSuperior]
dfOutliers
```

```python
#Como não existem Outliers, vou selecionar os 10 maiores

#Ordeno o Data Frame primeiro de forma decrescente
dfApreeMenores = dfApreeMenores.sort_values("aaapai", ascending=False)

#Crio o Data Frame com os 5 maiores casos
dfOutliers = dfApreeMenores.head(10)
dfOutliers
```

```python
#Exporto o DataFrame para um arquivo CSV
dfOutliers.to_csv("dfMunicOutliersApreeMenoresBaixadaFlu.csv", ";", index=False)
```
