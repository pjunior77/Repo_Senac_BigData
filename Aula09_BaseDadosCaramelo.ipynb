```python
import pandas as pd

df = pd.read_csv('CARAMELO.csv', delimiter=";")
```

```python
df.head(2)
```

```python
#Altera o campo "Valor de Venda" de string para float (pois estava dando erro ao calcular a soma)
df['Valor de Venda'] = df['Valor de Venda'].str.replace(',','.').astype(float)
df.head(2)
```

```python
#Conta as vendas realizadas agrupadas pela "Data da Compra"
df_DataCompra_QtdeVenda = df.groupby('Data da Compra')['ID Venda'].count().reset_index()

#Exibe as datas que tiveram mais vendas
df_DataCompra_QtdeVenda_DESC = df_DataCompra_QtdeVenda.sort_values(by='ID Venda', ascending=False)
df_DataCompra_QtdeVenda_DESC
```

```python
#Soma o "valor de venda" agrupado por "Produto"
df_Produto_ValorVenda = df.groupby('Produto')['Valor de Venda'].sum().reset_index()
df_Produto_ValorVenda
```

```python
#Soma o "valor de venda" agrupado por "Nome Cliente"
df_NomeCliente_ValorVenda = df.groupby('Nome Cliente')['Valor de Venda'].sum().reset_index()
df_NomeCliente_ValorVenda
```

```python
#Calcula o valor médio do "valor de venda" agrupado por "Sexo"
df_Sexo_ValorVenda = df.groupby('Sexo')['Valor de Venda'].mean().reset_index()
df_Sexo_ValorVenda
```

```python
#Calcula o valor médio do "valor de venda" agrupado por "Metodo Pagamento"
df_MetodoPagamento_ValorVenda = df.groupby('Metodo de Pagamento')['Valor de Venda'].mean().reset_index()
df_MetodoPagamento_ValorVenda
```

```python
#Soma o "valor de venda" agrupado por "Estado Loja"
df_EstadoLoja_ValorVenda = df.groupby('Estado Loja')['Valor de Venda'].sum().reset_index()
df_EstadoLoja_ValorVenda
```

```python
#Soma o "valor de venda" agrupado por "Categoria Produto"
df_CatProduto_ValorVenda = df.groupby('Categoria Produto')['Valor de Venda'].sum().reset_index()
df_CatProduto_ValorVenda
```

```python
#Conta as "lojas" agrupadas por "Categoria Produto"
df_CatProduto_QtdeLojas = df.groupby('Categoria Produto')['ID Loja'].count().reset_index()
df_CatProduto_QtdeLojas
```
