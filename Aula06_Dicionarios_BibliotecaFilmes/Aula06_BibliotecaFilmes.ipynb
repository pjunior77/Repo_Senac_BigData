### AULA 06 - Atividade Biblioteca de Filmes

```python
#Cria o dicionário (Chave: Usuário x Valor: Lista de Filmes)
usuarios_filmes = {}

while True:
    print('\n=== MENU ===')
    print('1 - ADICIONAR filme')
    print('2 - REMOVER filme')
    print('3 - Ver filmes de um usuário')
    print('4 - Ver todos os usuários')
    print('0 - SAIR')

    #Pede ao usuário para digitar uma opção
    while True:
        try:
            opcaoMenu = int(input('Qual a opção escolhida? '))
            break
        except:
            print('Valor inválido. Digite um número inteiro.')
    
    #Verifica qual a opção escolhida
    match opcaoMenu:

        #Adicionar filme --------------------
        case 1:
            #Solicita o nome do usuário e o filme
            nomeUsuario = input('Digite o nome do usuário: ').strip()
            nomeFilme = input('Digite o nome do filme que deseja incluir: ').strip()

            #Verifica se o usuário informado não existe entre as chaves do dicionário usuarios_filmes
            if nomeUsuario not in usuarios_filmes:

                #Usuário não existe, cria o usuário com uma lista de filmes vazia
                usuarios_filmes[nomeUsuario] = []

            #Verifica se o filme informado já existe na lista de filmes do usuário (chave)
            if nomeFilme not in usuarios_filmes[nomeUsuario]:

                #Filme não existe, inclui o filme informado na lista de filmes do usuário (chave)
                usuarios_filmes[nomeUsuario].append(nomeFilme)

                print('\nFilme adicionado com sucesso!')
            
            else:
                #Filme já existe, imprime mensagem que o filme já existe
                print(f'\nO filme "{nomeFilme}" já está cadastrado para o usuário "{nomeUsuario}".')


        #Remover um filme --------------------
        case 2:
            #Solicita o nome do usuário e o filme
            nomeUsuario = input('Digite o nome do usuário: ').strip()
            nomeFilme = input('Digite o nome do filme que deseja excluir: ').strip()

            #Verifica se o usuário informado existe entre as chaves do dicionário usuarios_filmes
            if nomeUsuario in usuarios_filmes:

                #Usuário existe, agora verifica se o filme existe na lista de filmes assistidos por ele
                if nomeFilme in usuarios_filmes[nomeUsuario]:

                    #Remove o filme da lista de filmes
                    usuarios_filmes[nomeUsuario].remove(nomeFilme)
                    print('\nFilme removido com sucesso.')
                
                else:
                    #Filme não encontrado
                    print('\nFilme não encontrado.')

            else:
                #Usuário não encontrado
                print('\nUsuário não encontrado.')


        #Listar os filmes assistidos por um usuário --------------------
        case 3:
            #Solicita o nome do usuário
            nomeUsuario = input('Digite o nome do usuário: ').strip()

            #Verifica se o usuário existe
            if nomeUsuario in usuarios_filmes:

                #Usuário existe, então lista os filmes assistidos por ele
                print(f'\nO usuário "{nomeUsuario}" já assistiu aos filmes: {usuarios_filmes.get(nomeUsuario)}')
            
            else:
                #Usuário não encontrado
                print('\nUsuário não encontrado.')


        #Listar os filmes assistidos por todos os usuários --------------------
        case 4:

            for nome, listaFilmes in usuarios_filmes.items():
                print(f'\nO usuário "{nome}" já assistiu aos filmes: {listaFilmes}')


        #Terminar o programa --------------------
        case 0:
            print('--- FIM ---')
            break


        #Exibir a mensagem de opção inválida --------------------
        case _:
            print('Opção inválida.')
        
```
