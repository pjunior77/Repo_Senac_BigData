```python
#Importo as bibliotecas
import pandas as pd
import numpy as np
```

```python
#Carrego a base de dados
dfGeral = pd.read_csv("https://www.ispdados.rj.gov.br/Arquivos/BaseDPEvolucaoMensalCisp.csv", sep=";", encoding="latin1")
dfGeral.head(2)
```

```python
#Seleciono os dados da Capital
dfCapital = dfGeral.loc[dfGeral["regiao"] == "Capital"]
dfCapital.head(2)
```

```python
#Soma os valores de estelionato (estelionato) por delegacia (cisp)
dfEstelionatoRJ = dfCapital.groupby("cisp")["estelionato"].sum().reset_index()
dfEstelionatoRJ.head(2)
```

```python
#Crio um array com os dados de estelionato
arrayEstelionatoRJ = np.array(dfEstelionatoRJ["estelionato"])
arrayEstelionatoRJ
```

```python
#Cálculos estatísticos
media = np.mean(arrayEstelionatoRJ)
mediana = np.median(arrayEstelionatoRJ)
q1 = np.percentile(arrayEstelionatoRJ, 25)
q2 = np.percentile(arrayEstelionatoRJ, 50)
q3 = np.percentile(arrayEstelionatoRJ, 75)
distancia = (media - mediana) / mediana
iqr = q3 - q1
limSuperior = q3 + (1.5 * iqr)
limInferior = q1 - (1.5 * iqr)

print(f'Média     = {media:.2f}')
print(f'Mediana   = {mediana:.2f}')
print(f'Q1        = {q1:.2f}')
print(f'Q2        = {q2:.2f}')
print(f'Q3        = {q3:.2f}')
print(f'Distância = {distancia * 100:.2f} %')       #Em percentual
print(f'Lim. Sup. = {limSuperior:.2f}')
print(f'Lim. Inf. = {limInferior:.2f}')
```

```python
#Seleciono as DPs com 25% maiores casos de estelionato do RJ
dfDP25MaisEstelionatoRJ = dfEstelionatoRJ.loc[dfEstelionatoRJ["estelionato"] >= q3]
dfDP25MaisEstelionatoRJ
```

```python
#Crio um arquivo CSV com os dados do Data Frame
dfDP25MaisEstelionatoRJ.to_csv('dfEstelionatoRJ25Mais.csv', ";", index=False)
```
