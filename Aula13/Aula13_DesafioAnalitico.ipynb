```python
#Importação das bibliotecas
import pandas as pd
import numpy as np
```

```python
#Importa os dados do arquivo CSV e monta o DataFrame com a Base de Dados
dfBD = pd.read_csv("https://www.ispdados.rj.gov.br/Arquivos/BaseDPEvolucaoMensalCisp.csv", sep= ";", encoding="latin1")
dfBD.head(2)
```

```python
#OBS: Não precisa disso, porque senão as delegacias sem furto não vão aparecer nos resultados
#Filtra os dados relacionados a furtos de celulares
#dfBD_furto_celulares = dfBD.loc[dfBD["furto_celular"] > 0]
#dfBD_furto_celulares.head(2)
```

```python
#Agrupa os dados por delegacia (cisp), somando o total de furtos registrados
df_FurtoCel_PorDelegacia = dfBD.groupby("cisp")["furto_celular"].sum().reset_index()
df_FurtoCel_PorDelegacia.head(2)
```

```python
#Cria o array com os dados do furto_celular
array_FurtoCel_PorDelegacia = np.array(df_FurtoCel_PorDelegacia["furto_celular"])
array_FurtoCel_PorDelegacia
```

```python
#Cálculo da média, mediana e quartis dos furtos
mediaFurto = np.mean(array_FurtoCel_PorDelegacia)
medianaFurto = np.median(array_FurtoCel_PorDelegacia)
q1Furto = np.percentile(array_FurtoCel_PorDelegacia, 25)
q2Furto = np.percentile(array_FurtoCel_PorDelegacia, 50)
q3Furto = np.percentile(array_FurtoCel_PorDelegacia, 75)
distancia = (mediaFurto - medianaFurto) / medianaFurto
iqrFurto = q3Furto - q1Furto
limiteSuperiorFurto = q3Furto + (1.5 * iqrFurto)
limiteInferiorFurto = q1Furto + (1.5 * iqrFurto)

print(f'Média     = {mediaFurto:.2f}')
print(f'Mediana   = {medianaFurto:.2f}')
print(f'Q1        = {q1Furto:.2f}')
print(f'Q2        = {q2Furto:.2f}')
print(f'Q3        = {q3Furto:.2f}')
print(f'Distância = {distancia:.2f}')             #Em número inteiro
print(f'Distância = {distancia * 100:.2f}')       #Em percentual
print(f'LimSup.   = {limiteSuperiorFurto:.2f}')
print(f'LimInf.   = {limiteInferiorFurto:.2f}')
```

```python
#Delegacias que representam os 25% com menor número de furtos (primeiro quartil)
delegacias_menorQ1 = df_FurtoCel_PorDelegacia.loc[df_FurtoCel_PorDelegacia["furto_celular"] <= q1Furto]
delegacias_menorQ1.sort_values(by="furto_celular")
```

```python
#Delegacias que representam os OUTLIERS com MENOR número de furtos
delegacias_MenorOutlier = df_FurtoCel_PorDelegacia.loc[df_FurtoCel_PorDelegacia["furto_celular"] <= limiteInferiorFurto]
delegacias_MenorOutlier.sort_values(by="furto_celular")
```

```python
#Delegacias que representam os 25% com maior número de furtos (terceiro quartil)
delegacias_maiorQ3 = df_FurtoCel_PorDelegacia.loc[df_FurtoCel_PorDelegacia["furto_celular"] >= q3Furto]
delegacias_maiorQ3.sort_values(by="furto_celular", ascending=False)
```

```python
#Delegacias que representam os OUTLIERS com MAIOR número de furtos
delegacias_MaiorOutlier = df_FurtoCel_PorDelegacia.loc[df_FurtoCel_PorDelegacia["furto_celular"] >= limiteSuperiorFurto]
delegacias_MaiorOutlier.sort_values(by="furto_celular")
```

```python
####################################################################################################################
#OBS: Tem que trocar a base com a visão por município, pois uma delegacia pode tomar conta de mais de um município!
####################################################################################################################
#Soma o total de estelionatos (estelionato) registrados por município (munic)
df_Estelionato_PorMunicipio = dfBD.groupby("munic")["estelionato"].sum().reset_index()
df_Estelionato_PorMunicipio.head(2)
```

```python
#Cria o array com os dados dos estelionatos
array_Estelionato_PorMunicipio = np.array(df_Estelionato_PorMunicipio["estelionato"])
array_Estelionato_PorMunicipio
```

```python
#Cálculo da média, mediana e quartis dos estelionatos
mediaEstelionato = np.mean(array_Estelionato_PorMunicipio)
medianaEstelionato = np.median(array_Estelionato_PorMunicipio)
q1Estelionato = np.percentile(array_Estelionato_PorMunicipio, 25)
q2Estelionato = np.percentile(array_Estelionato_PorMunicipio, 50)
q3Estelionato = np.percentile(array_Estelionato_PorMunicipio, 75)

print(f'Média   = {mediaEstelionato:.2f}')
print(f'Mediana = {medianaEstelionato:.2f}')
print(f'Q1      = {q1Estelionato:.2f}')
print(f'Q2      = {q2Estelionato:.2f}')
print(f'Q3      = {q3Estelionato:.2f}')
```

```python
#O objetivo é identificar aqueles municípios que representam os 25% que mais ocorreram tal fato
municipios_maiorQ3 = df_Estelionato_PorMunicipio.loc[df_Estelionato_PorMunicipio["estelionato"] >= q3Estelionato]
municipios_maiorQ3
```

```python
#Além disso, o Secretário quer identificar aqueles que menos ocorreram
municipios_menorQ1 = df_Estelionato_PorMunicipio.loc[df_Estelionato_PorMunicipio["estelionato"] <= q1Estelionato]
municipios_menorQ1
```

```python
#Soma o total de apreensão de menores (aaapai) registrados por município (munic)
df_ApreensaoMenores_PorMunicipio = dfBD.groupby("munic")["aaapai"].sum().reset_index()
df_ApreensaoMenores_PorMunicipio.head(2)
```

```python
#Cria o array com os dados das apreensões de menores
array_ApreensaoMenores_PorMunicipio = np.array(df_ApreensaoMenores_PorMunicipio["aaapai"])
array_ApreensaoMenores_PorMunicipio
```

```python
#Cálculo da média, mediana e quartis de apreensão dos menores
mediaApreensaoMenores = np.mean(array_ApreensaoMenores_PorMunicipio)
medianaApreensaoMenores = np.median(array_ApreensaoMenores_PorMunicipio)
q1ApreensaoMenores = np.percentile(array_ApreensaoMenores_PorMunicipio, 25)
q2ApreensaoMenores = np.percentile(array_ApreensaoMenores_PorMunicipio, 50)
q3ApreensaoMenores = np.percentile(array_ApreensaoMenores_PorMunicipio, 75)

print(f'Média   = {mediaApreensaoMenores:.2f}')
print(f'Mediana = {medianaApreensaoMenores:.2f}')
print(f'Q1      = {q1ApreensaoMenores:.2f}')
print(f'Q2      = {q2ApreensaoMenores:.2f}')
print(f'Q3      = {q3ApreensaoMenores:.2f}')
```

```python
#O objetivo é identificar aqueles municípios que representam os 25% que mais ocorreram apreensão de menores
municipiosApreeMenores_maiorQ3 = df_ApreensaoMenores_PorMunicipio.loc[df_ApreensaoMenores_PorMunicipio["aaapai"] >= q3ApreensaoMenores]
municipiosApreeMenores_maiorQ3
```

```python
#O objetivo é identificar aqueles municípios que representam os 25% que menos ocorreram apreensão de menores
municipiosApreeMenores_menorQ1 = df_ApreensaoMenores_PorMunicipio.loc[df_ApreensaoMenores_PorMunicipio["aaapai"] <= q1ApreensaoMenores]
municipiosApreeMenores_menorQ1
```
