```python
#Importa a biblioteca PANDAS
import pandas as pd

#Carrega o aruivo CSV
df = pd.read_csv('Precos_01sem_comb_autom_2025.csv', sep = ';', encoding = 'latin1')
df.head()
```

```python
df_RioDeJaneiro = df.loc[df['Municipio'] == 'RIO DE JANEIRO']
df_RioDeJaneiro.head()
```

```python
df_RioDeJaneiro_Etanol = df_RioDeJaneiro.loc[df['Produto'] == 'ETANOL']
df_RioDeJaneiro_Etanol.head()
```

```python
#Altera o campo de string para float (pois estava dando erro ao faer a média)
df_RioDeJaneiro_Etanol['Valor de Venda'] = df_RioDeJaneiro_Etanol['Valor de Venda'].str.replace(',','.').astype(float)
```

```python
#df_RioDeJaneiro_Etanol_Bairros = df_RioDeJaneiro_Etanol.groupby('Bairro')[('Valor de Venda')].sum().reset_index()
df_RioDeJaneiro_Etanol_Bairros = df_RioDeJaneiro_Etanol.groupby('Bairro')['Valor de Venda'].mean().reset_index()
```

```python
df_RioDeJaneiro_Etanol_Bairros_OrdenadoASC = df_RioDeJaneiro_Etanol_Bairros.sort_values(by='Valor de Venda')
df_RioDeJaneiro_Etanol_Bairros_OrdenadoASC.head()
```

```python
df_RioDeJaneiro_Etanol_Bairros_OrdenadoDESC = df_RioDeJaneiro_Etanol_Bairros.sort_values(by='Valor de Venda', ascending= False)
df_RioDeJaneiro_Etanol_Bairros_OrdenadoDESC.head()
```
