```python
#Importo as bibliotecas
import pandas as pd
import numpy as np
```

```python
#Carrego a base de dados
dfGeral = pd.read_csv("https://www.ispdados.rj.gov.br/Arquivos/BaseDPEvolucaoMensalCisp.csv", sep=";", encoding="latin1")
dfGeral.head(2)
```

```python
#Seleciono os dados da Baixada Fluminense
dfBaixFlu = dfGeral.loc[dfGeral["regiao"] == "Baixada Fluminense"]
dfBaixFlu.head(2)
```

```python
#Soma os valores de apreensão de menores por delegacia (aaapai) por delegacia (cisp)
dfApreeMenores = dfBaixFlu.groupby("cisp")["aaapai"].sum().reset_index()
dfApreeMenores.head(2)
```

```python
#Cria o array com os dados da apreensão de menores
arrayApreeMenores = np.array(dfApreeMenores["aaapai"])
arrayApreeMenores
```

```python
#Cálculos estatísticos
media = np.mean(arrayApreeMenores)
mediana = np.median(arrayApreeMenores)
q1 = np.percentile(arrayApreeMenores, 25)
q2 = np.percentile(arrayApreeMenores, 50)
q3 = np.percentile(arrayApreeMenores, 75)
distancia = (media - mediana) / mediana
```
