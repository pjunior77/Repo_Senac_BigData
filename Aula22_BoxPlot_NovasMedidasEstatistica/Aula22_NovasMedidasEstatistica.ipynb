```python
#Importa a biblioteca para cálculos estatísticos do Python
import numpy as np

#Importa a biblioteca do pandas
import pandas as pd

#Importa a biblioteca para gerar gráficos
import matplotlib.pyplot as plt
```

```python
#Cria o array com os dados
dados_array = np.array([5, 8, 12, 15, 20, 25, 30])

#Calcula os valores máximo, mínimo e a amplitude
minimo = np.min(dados_array)
maximo = np.max(dados_array)
amplitude = maximo - minimo

print(f'Máximo:    {maximo}\nMínimo:    {minimo}\nAmplitude: {amplitude}')
```

```python
import random

#Lista com os nomes dos vendedores
vendedores = [
    'Josefa',
    'Manoel',
    'Sérgio',
    'Francisca',
    'Teodoro',
    'Maria',
    'Vitor',
    'Vitória',
    'Pedro',
    'Teresa',
    'Luis',
    'Luisa',
    'Frederico',
    'Lais',
    'Marcos',
    'Manoela',
    'Henrique',
    'Florentina',
    'Ximbinha',
    'Sara',
    'Falcão',
    'Renata',
    'Chaves',
    'Chiquinha',
    'Kiko',
    'Dona Florinda',
    'Nhonho',
    'Bruxa do 71',
    'Mickey',
    'Minie',
]

#Lista com valores randômicos "random.dandint(valor mínimo, valor máximo)"
valores_vendidos = [random.randint(1000, 1000000) for i in vendedores]

#Adiciona dados Outliers as listas (porque o Random não garante que vai gerar valores com os seus limites)
vendedores.append('Tirulipa')
valores_vendidos.append(3000000)

#Cria o dicionário com os nomes dos vendedores e dos valores vendidos
dados_vendas = {'vendedores': vendedores,
                'valores_vendidos': valores_vendidos}

df = pd.DataFrame(dados_vendas)
df.sort_values(by='valores_vendidos', ascending=False, inplace=True)
df
```

```python
valores_vendidos_array = np.array(df['valores_vendidos'])
```

```python
plt.figure(figsize=(15,5))
plt.boxplot(valores_vendidos_array, vert=False, showmeans=True)
    #showmeans=True (para trazer a média)
    #vert=False (para trazer o gráfico na horizontal)
plt.title('BOXPLOT')
```

```python
import numpy as np

dados_array = np.array([2, 4, 6, 8, 10])
media = np.mean(dados_array)
variancia = np.var(dados_array)
desvio_padrao = np.std(dados_array)
coeficiente_variacao = (desvio_padrao / media) * 100
distancia_variancia = (variancia / (media ** 2)) * 100

print("Média", round(media, 2))
print("Variância:", round(variancia, 2))
print("Desvio padrão:", round(desvio_padrao, 2))
print("Coeficiente de variação:", round(coeficiente_variacao, 2))
print("Distância da variância em relação à média:", round(distancia_variancia, 2))
```
